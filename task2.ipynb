## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
## Reading the dataset

df=pd.read_csv('C:\\Users\\g_gup\\Desktop\\Iris (1).csv')

# Watching the dataset
df.head(2)
```

```python
df=df.iloc[:,1:-1].values
type(df)
```

```python
##use Elbow method To find the best k 

wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
# plot elbow curve
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Training and predicting the clusters
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(df)
y_kmeans
#plotting the clusters
plt.scatter(df[y_kmeans==0,0],df[y_kmeans==0,1],s=50,c='red',label='Iris-setosa')
plt.scatter(df[y_kmeans==1,0],df[y_kmeans==1,1],s=50,c='blue',label='Iris-versicolor')
plt.scatter(df[y_kmeans==2,0],df[y_kmeans==2,1],s=50,c='green',label='Iris-virginica')


plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='Centroids',marker='*')

plt.title('Clusters of Species')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.legend()
plt.show()
```
